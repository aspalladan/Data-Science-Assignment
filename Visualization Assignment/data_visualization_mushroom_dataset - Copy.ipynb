# Visualizing the Mushroom dataset

# Plotting the distribution of cap shape

```python

plt.figure(figsize=(8, 6))
df['cap-shape'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Cap Shape')
plt.xlabel('Cap Shape')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Plotting the distribution of cap color

```python

plt.figure(figsize=(8, 6))
df['cap-color'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Cap Color')
plt.xlabel('Cap Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Plotting the distribution of odor

```python

plt.figure(figsize=(8, 6))
df['odor'].value_counts().plot(kind='bar', color='salmon')
plt.title('Distribution of Odor')
plt.xlabel('Odor')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Plotting  the distribution of class

```python

plt.figure(figsize=(8, 6))
df['class'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Mushroom Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0, labels=['Edible', 'Poisonous'])
plt.show()
```

# Plotting pie charts for each categorical variable

```python


categorical_columns = df.select_dtypes(include=['object']).columns


for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Distribution of {column}')
    plt.ylabel('')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```
