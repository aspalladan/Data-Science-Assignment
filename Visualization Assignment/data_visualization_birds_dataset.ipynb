# Visualizing the bird dataset

```python

import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Read the data and display the first few rows
birds = pd.read_csv(r"C:\Users\Engr. CY\Desktop\ArewaDS\Data-Science-Assignment\Data preparation with the bird dataset\birds.csv")
birds.head()
```

```python
# Check the dimension of the dataset
birds.shape
```

```python
birds.info()
```

```python
birds.Name.nunique()
```

```python
wingspan = birds['MaxWingspan'] 
wingspan.plot()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.xticks(rotation=45)
x = birds['Name'] 
y = birds['MaxWingspan']

plt.plot(x, y)

plt.show()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x, y, 'bo')
    if birds['MaxWingspan'][i] > 500:
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=12)
    
plt.show()
```

```python
birds.plot(kind="line")
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    if birds['Name'][i] not in ['Bald eagle', 'Prairie falcon']:
        plt.plot(x, y, 'bo')
plt.show()
```
